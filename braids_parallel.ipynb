### Reference of (p,q) numbering system

[See List](https://en.wikipedia.org/wiki/Torus_knot#List)

```python
def torus_braids(p, q):
    if p < 3:
        raise ValueError("p must be >= 3")
        
    k = (p - 1) * q

    first_tuples = []
    second_tuples = []

    if p % 2 == 1:
        first = [j * q for j in range(0, p - 2, 2)]
        first.append((p - 2) * q)
        first = list(dict.fromkeys(first))

        second = [k - 1 - a for a in first]
        second.reverse()

        for x in first[:-1]:
            first_tuples.append((x, 3))
            first_tuples.append((x, 0))
        first_tuples.append((first[-1], 0))

        second_tuples.append((second[0], 2))
        for x in second[1:]:
            second_tuples.append((x, 2))
            second_tuples.append((x, 1))

    else:
        first = [j * q for j in range(0, p, 2)]
        first = list(dict.fromkeys(first))

        second = [q - 1] + [j * q - 1 for j in range(2, p, 2)] + [k - 1]
        second = list(dict.fromkeys(second))

        for x in first:
            first_tuples.append((x, 3))
            first_tuples.append((x, 0))

        second_tuples.append((second[0], 2))
        for x in second[1:-1]:
            second_tuples.append((x, 2))
            second_tuples.append((x, 1))
        second_tuples.append((second[-1], 1))

    connections = list(zip(second_tuples, first_tuples))

    return connections
```
